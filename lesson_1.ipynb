### 4. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()),если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
def replace_dict(text : pd.DataFrame, my_dict : dict) -> pd.DataFrame:
    
    text_new = str()
    for word in text.split():
        if word in my_dict.keys():
            word = my_dict.get(word)
        text_new = text_new + word+ ' '
        
    return text_new
```

```python
replace_dict_vectorized = np.vectorize(replace_dict)
combine_df['tweet_apostrophe'] = replace_dict_vectorized(combine_df['lower'], apostrophe_dict)
combine_df.head(10)
```

### 10 .Удалим из текста слова длиной в 1 символ, используя ' '.join([w for w in x.split() if len(w)>1])

```python
def del_small_token(combine_df: pd.DataFrame) -> pd.DataFrame :
    
    text_new = str()
    for word in combine_df.split():
        if len(word) <= 1:
            word = ' '
        text_new = text_new + word+ ' ' 
        
    return text_new
```

```python
del_small_token_vectorized = np.vectorize(del_small_token)
combine_df['tweet_drop_small_simbol'] = del_small_token_vectorized(combine_df['tweet_number'])
combine_df.head(10)
```

### 11. Поделим твиты на токены с помощью nltk.tokenize.word_tokenize, создав новый столбец 'tweet_token'.

```python
def tokenize (combine_df : pd.DataFrame) -> pd.DataFrame:
    
    combine_df['tweet_token'] = [nltk.tokenize.word_tokenize(w) for w in combine_df['tweet_drop_small_simbol']]
```

### 12. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.

```python
def drop_stopword(combine_df : pd.DataFrame) -> pd.DataFrame:
    
    stop_words = set(nltk.corpus.stopwords.words("english"))
    
    return [w for w in combine_df if not w in stop_words]
```

```python
combine_df['tweet_token_filtered'] = list(map(drop_stopword, combine_df['tweet_token']))
combine_df.head(10)
```

### 13.Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.

```python
def stemmer(combine_df : pd.DataFrame) -> pd.DataFrame:
    
    stemmer = nltk.stem.PorterStemmer()
    
    text_new = []
    for word in combine_df:
        word = stemmer.stem(word)
        text_new.append(word)
        
    return text_new
```

```python
combine_df['tweet_stemmed'] = list(map(stemmer, combine_df['tweet_token_filtered']))
combine_df.head(10)
```

### 13. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.

```python
def lemmatizer (combine_df : pd.DataFrame) -> pd.DataFrame:
    
    lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
    
    text_new = []
    for word in combine_df:
        word = lemmatizer.lemmatize(word)
        text_new.append(word)
        
    return text_new
```

```python
combine_df['tweet_lemitized'] = list(map(lemmatizer, combine_df['tweet_token_filtered']))
combine_df.head(10)
```
